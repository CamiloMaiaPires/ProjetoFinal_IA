```python
import pandas as pd
import requests
import zipfile
from sklearn.preprocessing import MinMaxScaler
```

## Normnalização dos dados

```python
ibge = pd.read_csv('../tratados/dados/ibge_drs.csv')
ibge.head(3)
```

*Casos e óbitos*

```python
scaler = MinMaxScaler()
for cod in ibge['Código IBGE']:
    dataset =  pd.read_csv('../tratados/casos_obitos/' + str(cod) + '.csv')
    colunas = dataset.iloc[:, 4:]
    normalizado = scaler.fit_transform(colunas)
    dataset.iloc[:, 4:] = normalizado
    dataset = dataset.drop(columns = 'Unnamed: 0')
    dataset.to_csv('./casos_obitos/' + str(cod) + '.csv')
```

```python
scaler = MinMaxScaler()
for cod in ibge['Código IBGE']:
    dataset =  pd.read_csv('../tratados/isolamento/cidades/' + str(cod) + '.csv')
    #dataset = dataset.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'])
    colunas = dataset['media_isolamento'].iloc
    normalizado = scaler.fit_transform(colunas)
    dataset.iloc['media_isolamento'] = normalizado
    #dataset = dataset.drop(columns = 'Unnamed: 0')
    dataset.to_csv('./casos_obitos/' + str(cod) + '.csv')
    break;
dataset
```
