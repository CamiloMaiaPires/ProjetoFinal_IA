```python
import pandas as pd
import requests
import zipfile
from sklearn.preprocessing import MinMaxScaler
```

## Normnalização dos dados

```python
ibge = pd.read_csv('../dados/ibge_drs.csv')
ibge.head(3)
```

*Casos e óbitos*

```python
scaler = MinMaxScaler()
for cod in ibge['Código IBGE']:
    dataset =  pd.read_csv('../casos_obitos/' + str(cod) + '.csv')
    colunas = dataset.iloc[:, 4:]
    normalizado = scaler.fit_transform(colunas)
    dataset.iloc[:, 4:] = normalizado
    dataset.to_csv('./casos_obitos/' + str(cod) + '.csv')
```
