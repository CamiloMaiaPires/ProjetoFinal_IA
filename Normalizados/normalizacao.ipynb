```python
import pandas as pd
import requests
import zipfile
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

## Normalização dos dados

```python
ibge = pd.read_csv('../tratados/dados/ibge_drs.csv')
ibge.head(3)
```

*Casos e óbitos*

```python
nomes = {'Município' : [] , 'Código IBGE': [], 'Populacao': [], 'Total 1ª Dose': [], 'Total 2ª Dose': [], 'Total Unica': [], 'Total Doses Aplicadas': [],
       'População Vacinada dose1/População Total': [],
       'População Vacinada/População Total': [], 'data': [], 'diagnostico_covid19': [],
       'obito': [], 'media_movel_casos': [], 'media_movel_obitos': [], 'ocupacao_leitos': [],
       'media_isolamento': []}

for cod in ibge['Código IBGE']:
    dataset =  pd.read_csv('../Finais/' + str(cod) + '.csv')
    if(not(dataset.empty)):
        for x in dataset.iloc():
            dataset.at[x.name, 'ocupacao_leitos'] = x['ocupacao_leitos'].replace(",",".")
            if(type(x['media_isolamento']) == str):
                dataset.at[x.name, 'media_isolamento'] = x['media_isolamento'].replace("%", "")
                
        scaler = MinMaxScaler()
        col = dataset.iloc[:,[4,5,6,7,8,9,11,12, 13,14, 15, 16, 17]]
        normalizado = pd.DataFrame(data=nomes)
        normalizado[['Total 1ª Dose','Total 2ª Dose','Total Unica','Total Doses Aplicadas','População Vacinada dose1/População Total','População Vacinada/População Total','diagnostico_covid19','obito',  'media_movel_casos','media_movel_obitos', 'ocupacao_leitos', 'media_isolamento' ,17]] = scaler.fit_transform(col)
        normalizado[['Município', 'Código IBGE', 'data','Populacao']] = dataset[['Município', 'Código IBGE', 'data', 'Populacao']]
        norma = normalizado.rename(columns={17:"Mean.R"})
        norma.to_csv('./dados/' + str(cod) + '.csv')
        
    else:
        print(cod)
```
