```python
import pandas as pd
import datetime
```

```python
def daterange(start_date, end_date):
    for i in range(int((end_date - start_date).days)):
        yield start_date + datetime.timedelta(i)
```

```python
#gerando csv unico
keys = {'Município': [], 'Código IBGE': [], 'Populacao': [], 'Total 1ª Dose': [], 'Total 2ª Dose': [], 'Total Unica': [], 'Total Doses Aplicadas': [], 'População Vacinada dose1/População Total': [], 'População Vacinada/População Total': [], 'data': []}
uniao = pd.DataFrame(data=keys)
a = 0
for data in daterange(datetime.datetime(2021,2,21), datetime.datetime.now()-datetime.timedelta(1)):
    d1 = pd.read_csv('./historico/' + data.strftime('%m') + '/' + data.strftime('%Y-%m-%d.csv'))
    d1 = d1.drop(columns=['distribuidas', 'Vacinas Aplicadas/Distribuidas', 'Unnamed: 0'])
    d1['data'] = data.strftime('%Y-%m-%d')
    uniao = uniao.append(d1)
    uniao = uniao.astype({'Código IBGE': 'int64', 'Populacao': 'int64', 'Total 1ª Dose': 'int64', 'Total 2ª Dose': 'int64', 'Total Unica': 'int64', 'Total Doses Aplicadas': 'int64'})
uniao
```

```python
cidades = uniao['Código IBGE'].unique()

for codigo in cidades:
    novo = pd.DataFrame(data=keys)
    cidade = uniao.groupby('Código IBGE').get_group(codigo)
    cidade = cidade.sort_values('data')
    cidade.to_csv('./vacinacao/' + str(codigo) + '.csv')
        
```
