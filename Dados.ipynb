```python
acento = pd.read_csv('Acento_Setado.csv').drop(columns=['Unnamed: 0','Total 1ª Dose', 'Total 2ª Dose', 'Total Doses Aplicadas', 'Doses distribuidas', '% Vacinas Aplicadas'])
acento = acento.sort_values('Código IBGE').reset_index().drop(columns='index')
acento
```

```python
ibge = pd.read_csv('IBGE.csv').drop(columns=['Unnamed: 0', 'Receitas_Realizadas', 'Escolarizacao_6_14_anos', 'IDHM'])
#inserir codigo ibge na tabela ibge
for i in ibge.iloc():
    municipio = i[0]
    if(not acento[acento['Município com acento'] == municipio].empty):
        codigo = acento[acento['Município com acento'] == municipio]['Código IBGE']
        ibge.at[i.name, 'Código IBGE'] = int(codigo.item())
        
    elif(not acento[acento['Municipio Sem Acento'] == municipio].empty):
        codigo = acento[acento['Municipio Sem Acento'] == municipio]['Código IBGE']
        ibge.at[i.name, 'Código IBGE'] = int(codigo.item())
    elif(not acento[acento['Municipio Sem Acento'] == unidecode(municipio)].empty):
        codigo = acento[acento['Municipio Sem Acento'] == unidecode(municipio)]['Código IBGE']
        ibge.at[i.name, 'Código IBGE'] = int(codigo.item())
    else:
        print(municipio)

ibge['Código IBGE'] = ibge['Código IBGE'].astype(int)
ibge = ibge.sort_values('Código IBGE').reset_index().drop(columns='index')
ibge
```

```python
cols = ['Indice s/ Acento', 'Indice c/ acento', 'Município com acento','Municipio Sem Acento']
ibge[cols] = acento[cols]
ibge
```

```python
for i in ibge.iloc():
    for drs in lista:
        for cidade in lista[drs]:
            if(unidecode(i[11]) == unidecode(cidade)):
                #print(drs, cidade)
                ibge.at[i.name, 'DRS'] = drs
ibge.head()
```
