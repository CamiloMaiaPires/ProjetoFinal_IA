# Casos e óbitos de covid confirmados por município do Estado de SP
### Obtenção dos dados

```python
import pandas as pd
import requests
import zipfile
```

```python
#fazendo download dos dados
co = requests.get('https://raw.githubusercontent.com/seade-R/dados-covid-sp/master/data/casos_obitos_doencas_preexistentes.csv.zip')
if (co.status_code == 200):
    open('casos_obitos_doencas_preexistentes.csv.zip', 'wb').write(co.content)
else:
    print('Link inválido.')
#descompactando o csv
with zipfile.ZipFile('casos_obitos_doencas_preexistentes.csv.zip', 'r') as zip_co:
    zip_co.extractall('./')
```

### limpeza e formatação

```python
casos = pd.read_csv('casos_obitos_doencas_preexistentes.csv', sep=';')
casos
```

```python
casos = casos.replace('CONFIRMADO', 1)
casos = casos.sort_values('data_inicio_sintomas')
casos
```

```python
casos = casos.dropna()
casos
```

```python
a = casos.groupby('nome_munic').get_group('Itapetininga')
a = a.groupby('data_inicio_sintomas').get_group('2020-02-29T00:00:00Z')
a
```

```python
cols = ['nome_munic', 'codigo_ibge', 'data_inicio_sintomas', 'diagnostico_covid19', 'obito']
keys = {'nome_munic': [], 'codigo_ibge': [], 'data_inicio_sintomas': [], 'diagnostico_covid19': [], 'obito': []}

cidades = casos['codigo_ibge'].unique()

for codigo in cidades:
    novo = pd.DataFrame(data=keys)
    cidade = casos[cols].groupby('codigo_ibge').get_group(codigo)
    datas = cidade['data_inicio_sintomas'].unique()
    for data in datas:
        dia = cidade.groupby('data_inicio_sintomas').get_group(data)
        diario = dia.head(1)[['nome_munic', 'codigo_ibge', 'data_inicio_sintomas']]
        diario['diagnostico_covid19'] = dia['diagnostico_covid19'].sum()
        diario['obito'] = dia['obito'].sum()
        novo = pd.merge(novo, diario, how='outer')
    print(novo.head(1)['nome_munic'][0])
    novo.to_csv('./casos_obitos/' + str(codigo) + '.csv')
        
```
