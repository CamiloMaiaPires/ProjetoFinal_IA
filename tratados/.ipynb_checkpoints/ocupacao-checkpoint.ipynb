# Dados sobre leitos no estado de São Paulo

#### Relacionar as DRS aos municípios.

```python
f = open('./dados/DRS.json', encoding="utf8")
lista = json.load(f)

ibge = pd.read_csv('./dados/ibge_drs.csv')
```

```python
ibge.head()
```

```python
ocupacao = pd.read_csv('./dados/plano_sp_leitos_internacoes_serie_nova_variacao_semanal.csv', sep=';').sort_values('nome_drs')
ocupacao.head()
```

```python
d = {'cod_ibge': [], 'data': [], 'ocupacao_leitos': []}
df = pd.DataFrame(data=d)
```

```python
#ocupacao = ocupacao.set_index("nome_drs")
#ocupacao = ocupacao.drop(
#    labels = ["Estado de São Paulo"], 
#    axis = 0, 
#    inplace = False
#)
#ocupacao = ocupacao.set_index("datahora")
```

```python
add = ['cod_ibge', 'data', 'ocupacao_leitos']
cities = []
for regiao in ocupacao.iloc():
    try:
        for city in ibge[ibge['DRS'] == regiao['nome_drs']].iloc():
            cities.append([city['Código IBGE'], regiao['datahora'], regiao['ocupacao_leitos_ultimo_dia']])
    except KeyError:
        pass
```

```python
df[add] = cities
```

```python
d = {'cod_ibge': [], 'data': [], 'ocupacao_leitos': []}

for id in ibge.iloc():
    csv = pd.DataFrame(data=d)
    csv = df[df['cod_ibge'] == id['Código IBGE']].sort_values(by='data')
    csv.to_csv('./ocupacao/' + str(id['Código IBGE']) + '.csv')
```
