```python
isolamento = pd.read_csv('./dados/20210720_isolamento.csv', sep=';')
isolamento
```

```python
ibge = pd.read_csv('./dados/ibge_drs.csv')
ibge.dropna()
ibge
```

```python
d = {'DRS': [], 'data': [], 'media_isolamento': [], 'cidade' : [], 'cod_ibge':[]}



soma = 0
cont = 0

add = ['DRS', 'data', 'media_isolamento', 'cidade', 'cod_ibge']

DRS = ["DRS 08 Franca", "DRS 01 Grande São Paulo", "DRS 12 Registro", "DRS 04 Baixada Santista"]
datas = (isolamento['Data'].unique())

for x in DRS:
    try:
        cidades = ibge.groupby(ibge.DRS).get_group(x)
        df = pd.DataFrame(data=d)
        lis = []
        #print(cidades)
        print(x)
        for data in datas:
            #print(data)
            soma = 0
            cont = 0
            media = 0
            for  dia_isolamento in isolamento.groupby(isolamento.Data).get_group(data).iloc():
                oi = []
                for city in cidades.iloc():
                    if(city['Código IBGE'] == dia_isolamento['Código Município IBGE']):
                        oi.append(city['Código IBGE'])
                       #print('----', dia_isolamento['Município1'], dia_isolamento['Média de Índice De Isolamento'], '----')
                        lis.append([city['DRS'], dia_isolamento['Data'], dia_isolamento['Média de Índice De Isolamento'], city['Municipio'], city['Código IBGE']])
                        cont +=1
                        soma += int( dia_isolamento['Média de Índice De Isolamento'][0:2])

            try:
                media = soma/cont
            except ZeroDivisionError:
                media = soma

            for city in cidades.iloc():
                if( not (city['Código IBGE'] in oi)):
                    lis.append([city['DRS'], data, media, city['Municipio'], city['Código IBGE']])
                   # print("---------",city['Municipio'], media,"--------");
                
        df[add] = lis
        df.to_csv("./isolamento/" + x + ".csv")
    except nan:
        pass
        
        
        
            
        #print("----", soma/cont, "--------")          
    
```

```python
for id in ibge.iloc():
    csv = isolamento[isolamento['Código Município IBGE'] == id['Código IBGE']].sort_values(by='Data')
    csv.to_csv('./isolamento/' + str(id['Código IBGE']) + '.csv')
```
